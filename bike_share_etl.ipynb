```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
import pymysql
pymysql.install_as_MySQLdb()
```

```python
weather_file = "Resources/weather.csv"
weather_df = pd.read_csv(weather_file)
weather_df.head()
```

```python
weather_clean_df = weather_df[['Date', 'Max_Temperature_F','Mean_Temperature_F','Min_TemperatureF','Max_Humidity','Mean_Humidity','Min_Humidity',
                               'Max_Wind_Speed_MPH','Mean_Wind_Speed_MPH','Max_Gust_Speed_MPH','Events']]

weather_clean_df_2 = weather_clean_df.rename(columns={'Min_TemperatureF':'Min_Temperature_F'})
weather_clean_df_2.set_index('Date',inplace=True)
weather_clean_df_2.head()
```

```python
trip_file = "Resources/trip_2.csv"
trip_df = pd.read_csv(trip_file,error_bad_lines=False, low_memory=False)
trip_df.set_index('trip_id',inplace=True)
trip_df.head()
```

```python
station_file = "Resources/station.csv"
station_df = pd.read_csv(station_file)
station_df_2 = station_df.rename(columns={'long':'lng'})
station_df_2.set_index('station_id',inplace=True)
station_df_2.head()
```

```python
connection_string = "root:" + password + "@127.0.0.1:3306/bikeshare_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
engine.table_names()
```

```python
weather_clean_df_2.to_sql(name='weather', con=engine, if_exists='append', index=True)
```

```python
station_df_2.to_sql(name='station', con=engine, if_exists='append', index=True)
```

```python
trip_df.to_sql(name='trip', con=engine, if_exists='append', index=True)
```
